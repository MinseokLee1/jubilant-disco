```python
# pandas, numpy 기초
# 유튜브 "파이스탁"
import numpy as np
import pandas as pd
```

```python
# numpy.
# 사실상 NdArray 구조를 사용하고자 하는 것, 이거 R에서의 자료구조와 비슷.

arr1 = np.array([1, 2, 3])
lis1 = [1, 2, 3] ; lis2 = []

# 파이썬 list에서 for 문으로 접근해야 하는 걸 NdArray에서는 그냥 연산이 됨.
for i in lis1:
    lis2.append(i * 10)

print(arr1 * 10, lis2)
```

```python
# 파이썬 리스트에 곱연산은 아래처럼 됨.
print(lis1 * 10)

# 참고. 튜플 () 은 elements 수정이 안 됨.
```

```python
# 다차원 데이터.

# 파이썬 리스트에서.
price = [
    [100, 80, 70, 90],
    [120, 110, 100, 110]
]

print(1, price) ; print()
print(2, price[0])
# 다만 이렇게 쓰면 column을 얻어올 순 없음. 행렬 구조가 아니기 때문.
```

```python
# ndarray에서.
price_arr = np.array(price)

print(1, price_arr); print()
print(2, price_arr[0]); print()
print(3, price_arr[0,]); print()
print(4, price_arr[:,-1]); print()
print(5, price_arr[0][0:2]); print()
print(6, price_arr[0,0:2])
#파이썬은 행렬간 경계선 기준으로 0, 1, ... 으로 인덱싱한다는 점에 유의.
```

```python
# pandas.
# Series : pd 1차원 자료구조.
arr = [0, 1, 2, 3]
ind = ["a", "b", "c", "d"]
s = pd.Series(data = arr, index = ind)
print(1) ; print(s); print()

print(2) ; print(s.index); print() # 딕셔너리의 key와 비슷한 역할.
print(3) ; print(s.values) ; print()
print(4) ; print(s.array)
```

```python
# Series는 array, index, row number로 되어 있음.
print(1) ; print(s.iloc[0:2]); print() # 행 번호 속성.
print(2); print(s.loc[["a", "c"]]) # 인덱스 속성.
```

```python
# Series 연산.
# 추가. 원본 바뀜.
s.loc["e"] = 4 ; print(s) # 딕셔너리 키-밸류 추가하는 느낌이라 여러 개 동시에 안 됨.
```

```python
# 삭제. 원본 유지.
print(1) ; print(s.drop(["d", "e"])) ; print()
print(2) ; print(s)
```

```python
# 수정. 원본 바뀜.
print(1) ; s.iloc[4] = 1 ; print(s); print()
print(2) ; s.loc["e"] = 4; print(s)
```

```python
# Broadcasting. 반복문 없이, 연산이 시리즈 전체 값에 적용한다는 개념.
# 아래를 보고 "브로드캐스팅 됐다" 라고 함.
print(s + 10)
```

```python
# 같은 index를 갖는 series들은 사칙연산 및 Boolean에 대해 브로드캐스팅이 됨.
# 순서와 상관 없이 인덱스를 기준으로 계산.
s1 = pd.Series(data = [10, 20, 30, 40, -50], index = ["a", "b", "c", "d", "e"])
s2 = pd.Series(data = [200, 100, 300], index = ["b", "a", "d"])

print(1) ; print(s1 - s) ; print()
print(2) ;print(s + s2) ; print()
print(3) ;print(s1 > s) ; print()
print(4) ;cond = s2 >= 200 ; print(cond) ; print()
print(5) ;print(s2[cond])
```

```python
# 정리.

# 종가가 80000원 이상 90000원 미만인 날짜 ?
dat = [93000, 82400, 99100, 81000, 72300]
ind = ["05/14", "05/15", "05/16", "05/17", "05/18"]
종가 = pd.Series(data = dat, index = ind)
print(1) ; print(종가) ; print()

# pd에서는 and 대신 &, or 대신 |, not 대신 ~을 씀. numpy에서 이걸 쓴다고 함.
# boolean을 여러개 쓸 때는 () 로 묶어줘야 함. 그랬을 때 cond는 bool series를 return.
cond = (종가 >= 80000) & (종가 < 90000)
print(2) ; print(종가[cond].index) ; print()

# for loop로 이쁘게 뽑는 것도 가능.
print(3)
for data in 종가[cond].index :
    print(data)
```

```python
# df col 및 row의 추가, 삭제.

# 샘플 데이터. 출처는 전공때 들은 회귀분석 강의.
df = pd.read_csv("C:/Users/lmseo/RWD/SPATCH/data/toy/TBC_dualphoton.csv")
df = df.rename(columns = {"Y=(%)Fat" : "Fat",
                          "X1=Age" : "Age",
                          "X2=Gender" : "Gender"})
print(df.shape)
df.head()
```

```python
# 일반적으로 series 단위 연산을 사용하므로 broadcasting이 됨.
# col 추가.
df["temp"] = df["Fat"] * df["Age"]
df.head()
```

```python
# 원본을 남기지 않으려면.
df.drop("temp", axis = 1, inplace = True)
df.head()

# Timeseries는 일단 잘 안쓸 것 같아서 패스.
# 쓸거면 index = pd.to_datatime( <string> ), 얘 타입은 imestamp 내지 DatatimeIndex. 이걸 인덱스로 사용.
```

```python
# 컬럼 시프트.
# 위의 브로드캐스팅 연산은 기본적으로 같은 index의 컬럼 사이에서 적용.
# 만약 내 아래 사람과 내 Fat 차를 보고 싶으면? 새로운 col을 만들어줘야 함.

# Fat 데이터를 아래로 한 칸씩 이동.
df["Fat_shift1"] = df["Fat"].shift(1)
df["Diff"] = df["Fat_shift1"] - df["Fat"]
df["Diff"] = df["Diff"].shift(-1)
df
```

```python
# 응용, 위 데이터의 normalization.
df.drop(["Fat_shift1", "Diff"], axis = 1, inplace = True)
df_norm = (df - df.mean()) / df.std()
df_norm.head()
```
