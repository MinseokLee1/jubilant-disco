```python
# matplotlib 베이직.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 샘플 데이터.
import seaborn as sns
titanic = sns.load_dataset("titanic") ; print(titanic.shape)
titanic.head()
```

```python
# 기본적으로는 하나씩 넣어주면 된다.
# 디폴트는 라인 그래프다.
# 아래는 그리기만 하고 오브젝트로 저장은 안 함.
plt.plot(titanic["age"], titanic["fare"])
plt.title("Titanic")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show() # 자잘한 것 빼고 플롯만 출력.
# plt.savefig( <filename> )
```

```python
# gridExtra 같은 것.

x = np.linspace(0, np.pi * 10, 500)
fig, axis = plt.subplots(2, 1) # gridExtra처럼 2row 1column으로 figure 넣을 백그라운드 생성.
# 위에서 object는 fig, axis는 ndarray. Indexing을 위한 것.
axis[0].plot(x, np.sin(x))
axis[1].plot(x, np.cos(x))
#fig.savefig()
```

```python
# ggplot2에서의 lty 혹은 type 등과 같은 것.

x = np.arange(-9, 10) #linspace와 비슷한데 얘는 length 말고 by를 받음.
y = x ** 2

plt.plot(x, y, linestyle = ":", marker = "*")
plt.show()
```

```python
# legend.

x = np.linspace(0, np.pi * 10, 500)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, linestyle = "-.", color = "red", label = "y = sin(x)", alpha = 0.5)
plt.plot(x, y2, linestyle = ":", color = "blue", label = "y = cos(x)", alpha = 0.5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("sin and cos")
plt.legend(shadow = False,
          borderpad = 1,
          loc = "upper right")
plt.show()
```

```python
# 다른 그래프 그리기.

# bar plot. 처음만 바꿔주면 된다.
plt.bar(x, y1, color = "green", label = "y = sin(x)", alpha = 0.1)
plt.bar(x, y2, color = "yellow", label = "y = cos(x)", alpha = 0.1)
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim([0, np.pi * 2])
plt.title("sin and cos")
plt.legend(shadow = False,
          borderpad = 1,
          loc = "upper right")
plt.show()
```

```python
# 누적 막대 그래프. 
# 함수가 따로 있는 건 아니고, 하나의 x에 대해 여러번 막대 그래프를 겹쳐 그리는 컨셉.
dat = [y1, y2]

for i in range(0, 2) : # y1, y2 두 종류.
    plt.bar(
        x, # x scale은 그대로.
        dat[i],
        bottom = np.sum(dat[:i], axis = 0), # y2는 y1 값에서 시작.
        alpha = 0.1
    )
    plt.xlim(0, np.pi * 2)
plt.show()
```

```python
# Scatter plot.
np.random.seed(123)

x = np.random.rand(10)
y = np.random.rand(10)

col = np.random.randint(0, 100, 10)  # 정수로 색상 지정 가능.
size = np.pi * 1e3 * np.random.rand(10)  # size도 10개 랜덤하게.

plt.scatter(x, y,
           c = col, #색깔
           s = size, # 크기
           alpha = 0.5)
plt.title("Fungi")  # 한글 입력 안됨
plt.show()
```
